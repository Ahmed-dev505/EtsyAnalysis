### Combine all Review Dates Fields  and Clean Text

```python
# takes a string
# returns a string

def clean_name(messy_name):
    new_list = []
    finder = re.compile("on " + "(.*?)" + "\r", re.IGNORECASE)
    matches = finder.findall(messy_name)
    return ";".join(matches)
```

```python
sampledf['clean_dates'] = ""
for i in range(0,1485): # i is index of shop in sampledf
    new_dates = []
    for n in range(1,9): # n is number of review pages in the columns of sampledf 
        if type(sampledf['review_dates' + str(n)][i]) == str:
            s = clean_name(sampledf['review_dates' + str(n)][i])
            new_dates.append(s)
        elif type(sampledf['review_dates' + str(n)][i]) == float:
            break
    sampledf['clean_dates'][i] = ";".join(new_dates)
```

### Define and Assign Seasons

| Season | Definition |
| --- | --- |
| Winter | December, January, February |
| Spring | March, April, May |
| Summer | June, July, August |
| Fall | September, October, November |
| oops | month not found |

```python
def get_month(date_str):
    x = date_str.split(" ")
    return x[0]
```

```python
reviewdf['review_month'] = reviewdf['review_date'].apply(get_month)
```

```python
def get_season(month_str):
    if month_str == "Dec" or month_str == "Jan" or month_str == "Feb":
        return "winter"
    elif month_str == "Mar" or month_str == "Apr" or month_str == "May":
        return "spring"
    elif month_str == "Jun" or month_str == "Jul" or month_str == "Aug":
        return "summer"
    elif month_str == "Sep" or month_str == "Oct" or month_str =="Nov":
        return "fall"
    else:
        return "oops"
```

```python
reviewdf['season'] = reviewdf['review_month'].apply(get_season)
```
